# Load indexes
This function load indexes data as dictionary in whitch, keys are indexes names and values are dataframes that contain informations Date,open,close,Hight,Low,Close,Adj Close,Volume

```python
dfDictist,indexes_names=load_indexes("../Data/indexes")
dfDictist["SSEC"].head()
```

```python
dfDictist["SSEC"]["Close"].values
```

### Plot with dates

```python
fig=plt.figure(figsize=(20,8))
plt.plot(dfDictist["SSEC"]["Close"])
plt.plot(dfDictist["SSEC"]["Open"])
fig.suptitle('Close value time series of SSEC index', fontsize=20)
plt.xlabel('Dates', fontsize=18)
plt.ylabel('Close', fontsize=16)
```

```python
rate=dfDictist["SSEC"].Close/dfDictist["SSEC"].Open
```

```python
fig=plt.figure(figsize=(20,8))
plt.plot(rate)
plt.suptitle("Open/Close rate for SSEC index", fontsize=20)
plt.xlabel('Dates', fontsize=18)
plt.ylabel('Rate', fontsize=16)
plt.show()
```
